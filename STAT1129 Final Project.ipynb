```python
import random
class Board:
    def __init__(self):
        self.win = 0
        self.even = 0
        self.lose = 0
        self.results = []
        self.score = 0
        self.mapping = {'rock': 0, 'paper': 1, 'scissors': 2}
        self.mapping_rev = ['rock', 'paper', 'scissors']
    
    def move(self):
        '''
        randomly generate a move by computer
        '''
        return random.randint(0, 2) 
    
    def validate(self, player_move):
        '''
        valid player input are strings of 'rock', 'paper', 'scissors'
        '''
        if player_move not in self.mapping:
            return -1
        return self.mapping[player_move] 
    
    def result(self, player_move):
        '''
        0 < 1, 1 < 2, 2 < 0
        '''
        pc_move = self.move()
        print('Your Move:', self.mapping_rev[player_move])
        print('Computer\'s Move:', self.mapping_rev[pc_move])
        if player_move == pc_move:
            print('It\'s Even This Round!')
            return 0
        elif (pc_move+1)%3 == player_move:
            print('You Win This Round!')
            return 1
        else:
            print('Computer Wins This Round!')
            return -1
    
    def scorekeeper(self, result):
        if result == 0:
            self.even += 1
            self.results.append('Even')
        elif result == 1:
            self.win += 1
            self.results.append('Win')
            self.score += 1
        else:
            self.lose += 1
            self.results.append('Lose')
            self.score -= 1
            
    def record(self):
        print('You have played {} game{} in total'.format(len(self.results), 's' if len(self.results) > 1 else ''))
        print('Current Score:', self.score)
        print('You have won', self.win, 'game{}'.format('s' if self.win > 1 else '') )
        print('Match Record:', self.results)
        
        
def main():
    print('Hi, Welcome to Our STAT1129 Final Project - Rock, Paper, Scissors Program')
    board = Board()
    while True:
        print('='*80)
        cmd = input('ENTER YOUR MOVE\n').lower()
        if cmd == 'exit':
            print('Exiting the program...')
            print('Here is the your record vs our program')
            board.record()
            break
        elif cmd == 'check':
            board.record()
        else:
            player_move = board.validate(cmd)
            if player_move == -1:
                print('Invalid Command, try again')
                continue
            result = board.result(player_move)
            board.scorekeeper(result)

if __name__ == '__main__':
    main()

main()
```
