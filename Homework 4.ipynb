```python
#Question 4

import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([2,3,4,5])
y2 = np.array([6,7,8,9])
y3 = np.array([10,11,12,13])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.xlabel('X Label')
plt.ylabel('Y Label')

plt.show()
```

```python
# Question 5

#takes three arguments that describes the layout of the figure
# the layout is organized in rows and columns, which are the first 2 values
# ie 1 row, 2 columns
#third argument represents the index of the current plot

import matplotlib.pyplot as plt
import numpy as np

#plot 1
x = np.array([0,1,2,3,4,5])
y = np.array([0,1,2,4,8,16])

plt.subplot(2,3,1)
plt.plot(x,y)
plt.title('hello')

#plot 2
x = np.array([0,1,2,3,4,5])
y = np.array([1,3,6,12,24,42])

plt.subplot(2,3,2)
plt.plot(x,y)
plt.title('heellloooo')

#plot 3
x = np.array([0,1,2,3,4,5])
y = np.array([1,1,2,5,8,9])

plt.subplot(2,3,3)
plt.plot(x,y)
plt.title('good morning')

#plot 4
x = np.array([0,1,2,3,4,5])
y = np.array([20,10,5,2.5,1,0])

plt.subplot(2,3,4)
plt.plot(x,y)
plt.title('hi!')

#plot 5
x = np.array([0,1,2,3,4,5])
y = np.array([1,4,2,6,4,5])

plt.subplot(2,3,5)
plt.plot(x,y)
plt.title('howdy')

#plot 6
x = np.array([0,1,2,3,4,5])
y = np.array([0,1,2,4,8,16])

plt.subplot(2,3,6)
plt.plot(x,y)
plt.title('good afternoon')

plt.suptitle('Greetings and lines')
plt.show()
```
